再写一个脚本处理这个结果文件
提取并排列成这个顺序：
display_name，first_name	middle_name（添加的列）	last_name	phone	is_phone_verified（添加的列）	email	is_email_verified（添加的列）	password（添加的列）	street_address_1 street_address_2	city	 region country postal_code date_joined

```python
import pandas as pd

def prepare_emails(
    infile: str = "merged_emails_output.xlsx",
    outfile: str = "final_emails_template.xlsx"
) -> None:
    """
    读取合并结果文件，抽取 / 新增指定列并按给定顺序排列。

    Parameters
    ----------
    infile : str
        输入 Excel 文件名（已包含合并后所有字段）。
    outfile : str
        输出 Excel 文件名。
    """
    # 读入原文件
    df = pd.read_excel(infile)

    # 新增缺失列并赋默认值
    df["middle_name"] = ""                # 默认空
    df["is_phone_verified"] = False       # 默认未验证
    df["is_email_verified"] = False       # 默认未验证
    df["password"] = ""                   # 默认空，可后续批量填充

    # 目标列顺序
    ordered_cols = [
        "display_name",
        "name",
        "first_name",
        "middle_name",
        "last_name",
        "phone",
        "is_phone_verified",
        "email",
        "is_email_verified",
        "password",
        "street_address_1",
        "street_address_2",
        "city",
        "region",
        "country",
        "postal_code",
        "date_joined",
    ]

    # 如果有列名大小写差异，可在此统一处理；假设列名已正确
    missing = [c for c in ordered_cols if c not in df.columns]
    if missing:
        raise KeyError(f"下列列缺失：{missing}")

    # 重新排序
    df_final = df[ordered_cols]

    # 保存结果（如不需要文件，可注释掉）
    df_final.to_excel(outfile, index=False)
    print(f"✅ Done! 结果已保存到 {outfile}")

if __name__ == "__main__":
    prepare_emails()
```
